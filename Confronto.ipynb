This notebook compares the amplitudes and the integrals between the two datasets

```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
ampiezze_min_on      = np.load('../Picosec/file/ampiezze_min_led_on.npy')
integrali_totale_on  = np.load('../Picosec/file/integrali_totali_led_on.npy') 
integrali_trigger_on = np.load('../Picosec/file/integrali_trigger_led_on.npy')

ampiezze_min_off      = np.load('../Picosec/file/ampiezze_min_led_off.npy')
integrali_totale_off  = np.load('../Picosec/file/integrali_totali_led_off.npy')
integrali_trigger_off = np.load('../Picosec/file/integrali_trigger_led_off.npy')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.hist(ampiezze_min_off, bins=200, density=True, color='darkblue', alpha=0.6, label='LED off')
ax.hist(ampiezze_min_on, bins=200, density=True, color='crimson',   alpha=0.5, label='LED on')

ax.set_xlabel('Ampiezza minima [V]')
ax.set_ylabel('Conteggi')
ax.set_title('Confronto distribuzioni delle ampiezze minime — LED off vs on')
ax.set_yscale('log') 

plt.legend()
plt.tight_layout()
plt.savefig('../outputs/confronto_hist_ampiezze.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.hist(integrali_totale_off, bins=200, density=True, color='darkblue', alpha=0.6, label='LED off')
ax.hist(integrali_totale_on, bins=200, density=True, color='crimson',   alpha=0.5, label='LED on')

ax.set_xlabel('Ampiezza minima [V]')
ax.set_ylabel('Conteggi')
ax.set_title('Confronto distribuzioni degli integrali totali — LED off vs on')
ax.set_yscale('log') 

plt.legend()
plt.tight_layout()
plt.savefig('../outputs/confronto_hist_integrali.png', dpi=150, bbox_inches='tight')
plt.show()
```
